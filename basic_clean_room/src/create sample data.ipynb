```python
from pyspark.sql import Row

# Fact table: patient_visits
patient_visits = [
    Row(visit_id=1, patient_id=101, doctor_id=201, hospital_id=301, diagnosis_id=401, visit_date="2025-12-01"),
    Row(visit_id=2, patient_id=102, doctor_id=202, hospital_id=302, diagnosis_id=402, visit_date="2025-12-02"),
    Row(visit_id=3, patient_id=103, doctor_id=201, hospital_id=301, diagnosis_id=403, visit_date="2025-12-03"),
    Row(visit_id=4, patient_id=104, doctor_id=203, hospital_id=303, diagnosis_id=404, visit_date="2025-12-04"),
    Row(visit_id=5, patient_id=101, doctor_id=202, hospital_id=302, diagnosis_id=401, visit_date="2025-12-05"),
]
df_patient_visits = spark.createDataFrame(patient_visits)

# Dimension table: patients
patients = [
    Row(patient_id=101, name="Alice Smith", dob="1980-01-01", gender="F"),
    Row(patient_id=102, name="Bob Jones", dob="1975-05-12", gender="M"),
    Row(patient_id=103, name="Carol Lee", dob="1990-07-23", gender="F"),
    Row(patient_id=104, name="David Kim", dob="1985-11-30", gender="M"),
]
df_patients = spark.createDataFrame(patients)

# Dimension table: doctors
doctors = [
    Row(doctor_id=201, name="Dr. Adams", specialty="Cardiology"),
    Row(doctor_id=202, name="Dr. Baker", specialty="Neurology"),
    Row(doctor_id=203, name="Dr. Clark", specialty="Pediatrics"),
]
df_doctors = spark.createDataFrame(doctors)

# Dimension table: hospitals
hospitals = [
    Row(hospital_id=301, name="General Hospital", city="Springfield"),
    Row(hospital_id=302, name="City Medical Center", city="Rivertown"),
    Row(hospital_id=303, name="Children's Hospital", city="Lakeside"),
]
df_hospitals = spark.createDataFrame(hospitals)

# Dimension table: diagnoses
diagnoses = [
    Row(diagnosis_id=401, code="I10", description="Hypertension"),
    Row(diagnosis_id=402, code="E11", description="Type 2 Diabetes"),
    Row(diagnosis_id=403, code="J45", description="Asthma"),
    Row(diagnosis_id=404, code="F32", description="Depression"),
]
df_diagnoses = spark.createDataFrame(diagnoses)

display(df_patient_visits)
display(df_patients)
display(df_doctors)
display(df_hospitals)
display(df_diagnoses)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS cr_owner_catalog.patient_visits")
```

```python
# Save each DataFrame as a Unity Catalog managed table
# Using default mode (errorifexists) to prevent accidental overwrites
df_patient_visits.write.saveAsTable("cr_owner_catalog.patient_visits.visits")
df_patients.write.saveAsTable("cr_owner_catalog.patient_visits.patients")
df_doctors.write.saveAsTable("cr_owner_catalog.patient_visits.doctors")
df_hospitals.write.saveAsTable("cr_owner_catalog.patient_visits.hospitals")
df_diagnoses.write.saveAsTable("cr_owner_catalog.patient_visits.diagnoses")

print("All tables created successfully in cr_owner_catalog.patient_visits schema")
```
